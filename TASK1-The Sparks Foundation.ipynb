# TASK 1 : PREDICTION USING SUPERVISED ML

```python
import pandas as pd
import matplotlib.pyplot as plt

#upload data
url = "http://bit.ly/w-data"
data = pd.read_csv(url)
#display first 5 rows
data.head()
```

```python

#plotting the distribution of scores
data.plot(x='Hours', y='Scores' , style='o')
plt.title('distribution of scores and hours')
plt.xlabel('Hours')
plt.ylabel('Scores')
plt.show()
```

```python
#splitting data to attributes and labels
X = data.iloc[:,:-1].values
y = data.iloc[:,1].values

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2, random_state=0)
```

```python
#Training The Algorithm
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(X_train,y_train)

print("training completed")
```

```python
#LinearRegression Line
line = reg.coef_*X+reg.intercept_

plt.scatter(X,y)
plt.plot(X,line)
plt.show()
```

```python
#Making Predictions
print(X_test)
pred = reg.predict(X_test)
```

```python
#Comparison of the actual 
df = pd.DataFrame({'Actual':y_test, 'Predicted' : pred})
df
```

```python
# The predited score if a student studies 9.25h/day

h = 9.25
h_pred = reg.predict([[h]])
print("The predicted score for "+str(h)+" Hours is "+str(h_pred[0]))
```

```python
#Evalluation of the model 
from sklearn import metrics  
print('Mean Absolute Error:', 
      metrics.mean_absolute_error(y_test, pred)) 
```
